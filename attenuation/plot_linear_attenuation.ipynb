```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

name = 'POND_21.png'
folder = r'D:\OneDrive - University of Victoria\Research\Attenuation Data\Txt Attenuation'

# List the file names and their densities, respectively
file_names = ['5pAu.txt', 'PEG.txt', 'H2O.txt']
density = [1.915, 1.21, 1]

# Make the legend
leg = ['5% Au', 'PEG', r'H$_2$O']

num_files = len(file_names)

sns.set_style('whitegrid')
fig = plt.figure(figsize=(7, 5))
# Plot each of the attenuation files
for i in range(num_files):
    data = np.loadtxt(os.path.join(folder, file_names[i]), dtype='float')
    energy = data[:, 0] * 1000  # Change to keV
    atten = data[:, 1] * density[i]  # Correct for density (linear attenuation)

    plt.semilogy(energy, atten, lw=2)

plt.xlabel('Energy (keV)', fontsize=18)
plt.ylabel(r'Linear attenuation coefficient (cm$^{-1}$)', fontsize=18)
plt.legend(leg, fontsize=18)
plt.tick_params(labelsize=15)
plt.xlim([1, 200])

plt.plot()
plt.savefig(os.path.join(folder, name), dpi=500)
```

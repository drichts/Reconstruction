```python
import os
import numpy as np
import matplotlib.pyplot as plt

directory = r'D:\OneDrive - University of Victoria\Research\LDA Data'
folder = '21-04-14_CT_bin_width_10'

bins = [70, 80, 90, 100, 110, 120]

subm = 'metal_phantom'
subw = 'water_phantom'

z = 13

# data = np.load(os.path.join(directory, folder, subm, 'Norm CT', 'CT_norm.npy'))
data = np.load(os.path.join(directory, folder, subw, 'Norm CT', 'CT_norm.npy'))

fig, ax = plt.subplots(2, 3, figsize=(10, 10))
ax[0, 0].axis('off')
ax[0, 1].axis('off')
ax[0, 2].axis('off')
ax[1, 0].axis('off')
ax[1, 1].axis('off')
ax[1, 2].axis('off')

ax[0, 0].imshow(data[0, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[0, 0].set_title(f'{bins[0]}-{bins[1]} keV')

ax[0, 1].imshow(data[1, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[0, 1].set_title(f'{bins[1]}-{bins[2]} keV')

ax[0, 2].imshow(data[2, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[0, 2].set_title(f'{bins[2]}-{bins[3]} keV')

ax[1, 0].imshow(data[3, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[1, 0].set_title(f'{bins[3]}-{bins[4]} keV')

ax[1, 1].imshow(data[4, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[1, 1].set_title(f'{bins[4]}-{bins[5]} keV')

ax[1, 2].imshow(data[5, z, 138:440, 138:440], cmap='gray', vmin=-400, vmax=400)
ax[1, 2].set_title('16-120 keV')
fig.show()
# fig.savefig(os.path.join(directory, folder, subw+f'_{z}.png'), dpi=500)
# fig.savefig(os.path.join(directory, folder, subm+f'_{z}.png'), dpi=500)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

directory = r'D:\OneDrive - University of Victoria\Research\LDA Data'
load_folder = '21-05-05_CT_metal_artifact'
save_dir = '21-05-05_CT_metal_30keV_bins'

sub_folders = ['water_only', '2_metal_1and2', '2_metal_3and4', '3_metal']

air_dark = ['airscan_60s', 'darkscan_60s']

for sf in sub_folders:
    data = np.load(os.path.join(directory, load_folder, sf, 'Data', 'data.npy'))
    save_folder = os.path.join(directory, save_dir, sf, 'Data')
    os.makedirs(save_folder, exist_ok=True)

    shp = np.shape(data)
    save_shp = np.copy(shp)
    save_shp[-1] = 3

    save_data = np.zeros(save_shp)
    for i in range(save_shp[-1]):
        save_data[:, :, :, i] = data[:, :, :, i] + data[:, :, :, i+1] + data[:, :, :, i+2]
        print(np.median(data[:, :, :, i]) + np.median(data[:, :, :, i+1]) + np.median(data[:, :, :, i+2]))
        print(np.median(save_data[:, :, :, i]))
        print()
    np.save(os.path.join(save_folder, 'data.npy'), save_data)

for af in air_dark:
    data = np.load(os.path.join(directory, load_folder, af, 'Data', 'data.npy'))
    save_folder = os.path.join(directory, save_dir, af, 'Data')
    os.makedirs(save_folder, exist_ok=True)

    shp = np.shape(data)
    save_shp = np.copy(shp)
    save_shp[-1] = 3

    save_data = np.zeros(save_shp)
    for i in range(save_shp[-1]):
        save_data[:, :, i] = data[:, :, i] + data[:, :, i+1] + data[:, :, i+2]
        print(np.median(data[:, :, i]) + np.median(data[:, :, i+1]) + np.median(data[:, :, i+2]))
        print(np.median(save_data[:, :, i]))
        print()

    np.save(os.path.join(save_folder, 'data.npy'), save_data)
```

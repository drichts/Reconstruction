```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def sigmoid(xpts):
    return 100 / (1 + np.exp(-1*xpts))

xpts1 = np.linspace(-10, 10, 100)
# xpts2 = np.linspace(40, 60, 100)

sns.set_style('whitegrid')

fig1 = plt.figure(figsize=(6, 4))
plt.plot(xpts1+20, sigmoid(xpts1), color='darkblue', ls='--')
plt.plot(xpts1+22, sigmoid(xpts1), color='orangered', ls='--')
plt.plot(xpts1+15, sigmoid(xpts1), color='darkblue', ls='-')
plt.plot(xpts1+27, sigmoid(xpts1), color='orangered', ls='-')
plt.arrow(19.5, 50, -4, 0, width=1, length_includes_head=True, head_length=1, color='darkblue')
plt.arrow(22.5, 50, 4, 0, width=1, length_includes_head=True, head_length=1, color='orangered')
plt.annotate('Tumour control', (0.08, 0.65), xycoords='axes fraction', va='center', fontsize=12)
plt.annotate('Adverse normal\n  tissue effects', (0.7, 0.34), xycoords='axes fraction', va='center', fontsize=12)
plt.xlabel('Total dose administered', fontsize=14)
plt.ylabel('Percent response', fontsize=14)
plt.title('Nanoparticle effect on the therapeutic index', fontsize=14)

# plt.legend(['Tumour control before', 'Adverse effect before', 'Tumour control after', 'Adverse effect after'])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
plt.tick_params(axis='y', labelsize=12)
plt.show()
# fig1.savefig(r'D:\OneDrive - University of Victoria\Files\Grad School\Writings and Presentations\Posters\WIMP21\t_index.png', dpi=500)
```
